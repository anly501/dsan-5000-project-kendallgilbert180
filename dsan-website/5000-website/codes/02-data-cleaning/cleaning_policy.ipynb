## Cleaning 7th Variable for Naive Bayes

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Loading Merged Data and New data
merged = pd.read_csv("/Users/kendallgilbert/dsan-5000-project-kendallgilbert180/dsan-website/5000-website/data/01-modified-data/merged.csv")
pi_nb = pd.read_csv("/Users/kendallgilbert/dsan-5000-project-kendallgilbert180/dsan-website/5000-website/data/00-raw-data/naive_bayes_policy.csv")
```

```python
new_header = pi_nb.iloc[0] #Grabbing the first row for the header
pi_nb = pi_nb[1:] #Taking the data less the header row
pi_nb.columns = new_header #Setting the header row as the df header
pi_nb = pi_nb.rename(columns={'Year': 'Country'})
pi_nb
```

```python
#Melting Data
melted_policy = pd.melt(pi_nb, id_vars=['Country'], var_name='Year', value_name='Value')

# Convert the "Year" column to a string and then combine "Country" and "Year" into "Country_Year"
melted_policy['Country_Year'] = melted_policy['Country'] + '_' + melted_policy['Year'].astype(str)

# Drop the original "Country" and "Year" columns
melted_policy = melted_policy.drop(columns=['Country', 'Year'])

# Convert the "Year" column to a string and then combine "Country" and "Year" into "Country_Year"
merged['Country_Year'] = merged['Country'] + '_' + merged['Year'].astype(str)

# Drop the original "Country" and "Year" columns
merged = merged.drop(columns=['Country', 'Year'])

nb_record = pd.merge(merged, melted_policy, on='Country_Year', how='inner')
```

```python
import numpy as np
# Split 'Country_Year' into 'Country' and 'Year'
nb_record[['Country', 'Year']] = nb_record['Country_Year'].str.split('_', expand=True)
# Drop the original 'Country_Year' column
nb_record = nb_record.drop('Country_Year', axis=1)
```

```python
#Exporting as CSV
nb_record.to_csv("nb_record.csv", header=True, index=False)
```
